```python
f=open('f1.txt','w')
f.write('some data into file')
f.close()
# i want to write the data again into file.
f.write("Some more data")
```

```python
f=open('f1.txt','w')
f.write('some data into file')
f.flush()
# i want to write the data again into file.
f.write("Some more data")
```

```python
l=['str1','str2','str3','str4']
f=open('f1.txt','w')
for i in l:
    f.write(i)
f.close()
```

```python
l=['str1','str2','str3','str4']
f=open('f1.txt','w')
f.writelines(l)
f.close()
```

```python
l=['str1\n','str2\n','str3\n','str4\n']
f=open('f1.txt','w')
f.writelines(l)
f.close()
```
